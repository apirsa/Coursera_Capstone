<h1> Toronto data </h1>   
Scraping the Wikipedia page

```python
import pandas as pd
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

Remove rows with borough "Not assigned"

```python
toronto_data=df[0]
toronto_data=toronto_data[toronto_data['Borough']!='Not assigned'].sort_values(by=['Postcode'])
toronto_data
```

Create list of unique postcodes:

```python
toronto_data
post_codes=toronto_data['Postcode'].drop_duplicates().to_list()
post_codes


```

For each value of post code, find part of dataframe with that post code. For each post code find unique values of neighbourhood and concate them into one string. If there is Not assigned value for neighbourhood, value of neighbourhood would be value of borough. Create dataframe with postcode, borough and joined values of neighbourhood. Append all dataframes in one final dataframe with right names of columns. 

```python
data=pd.DataFrame()
for i in post_codes:
    df=toronto_data[toronto_data['Postcode']==i]
    nei=""
    br=0
    for j in df['Neighbourhood'].drop_duplicates():
        if j=='Not assigned':
            j=df['Borough'].to_list()[0]
        if br>0:
            nei+= ", "+j
            br+=1
        else:
            nei+=j
            br+=1

    tmp=pd.DataFrame()
    data=data.append(pd.DataFrame({'Postcode':[i],'Borough':[df['Borough'].to_list()[0]], 'Neighbourhood':[nei]}))


toronto_data=data.reset_index()[['Postcode','Borough','Neighbourhood']]
```

```python
toronto_data.rename(columns={'Postcode':'PostalCode'}, inplace=True)
toronto_data
```
